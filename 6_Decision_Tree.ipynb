Task 6: Decision Tree Algorithm

A model classifies different species of Iris plant using Decision Tree Alogrithm

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Iris.csv")
df.head()
```

```python
df.pop("Id")
df.pop("Species")
df.head()
```

```python
features = df.iloc[:, [0, 1, 2, 3]].values
features
```

```python
import sklearn.datasets as dt

iris = dt.load_iris()

y = iris.target
y
```

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(df, y)
```

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

```python
dot_data = StringIO()
export_graphviz(dtree, out_file = dot_data, feature_names = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"], filled = True, rounded = True, special_characters = True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
